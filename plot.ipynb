# plot

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x = np.arange(20)
y = x**2

plt.plot(x, y)
```

```python
x = np.arange(-10,10)
y1 = 1/(1+np.exp(x))
y2 = 1/(1+np.exp(-x))

plt.subplot(1,2,1)
plt.plot(x, y1)
plt.subplot(1,2,2)
plt.plot(x, y2)
```

```python
x = np.arange(-2, 2, 0.01)
y1 = np.sin(np.pi*x)
plt.figure(figsize=(15,5))
y2 = np.sin(2*np.pi*x)
y3 = y1+y2
plt.subplot(1,3,1)
plt.plot(x, y1, label="y1")
plt.grid(True)
plt.legend(loc=1)
plt.subplot(1,3,2)
plt.plot(x, y2, label="y2")
plt.grid(True)
plt.legend(loc=1)
plt.subplot(1,3,3)
plt.plot(x, y3, label="y3")
plt.grid(True)
plt.legend(loc=1)
```
