# plot

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x = np.arange(20)
y = x**2

plt.plot(x, y)
```

```python
x = np.arange(-10,10)
y1 = 1/(1+np.exp(x))
y2 = 1/(1+np.exp(-x))

plt.subplot(1,2,1)
plt.plot(x, y1)
plt.subplot(1,2,2)
plt.plot(x, y2)
```
