```python
file_list = os.listdir(train_folder)
re_digits = re.compile(r'(\d+)')  
def emb_numbers(s):
    pieces=re_digits.split(s)
    pieces[1::2]=map(int,pieces[1::2])   
    return pieces
file_list.sort(key=emb_numbers)
# print(file_list)
```

```python
def get_name(index, hdf5_data):
    name = hdf5_data['/digitStruct/name']
    print(''.join([chr(v[0]) for v in hdf5_data[name[index][0]]]))

def get_label(index, hdf5_data):
    bbox = hdf5_data['/digitStruct/bbox']
    label = hdf5_data[bbox[index][0]]["label"]
    slabel = [0.0]*label.shape[0]
    for i in range(len(slabel)):
        slabel[i] = hdf5_data[label[i][0]][0][0]
    print(slabel)
for j in range(5):
    get_name(j, labels_file)
for j in range(5):
    get_label(j, labels_file)
```

 ## notMNIST 数据集拼接

读取pickle文件中的序列化对象

```python
pickle_file = '../notMNIST.pickle'

with open(pickle_file, 'rb') as f:
  save = pickle.load(f)
  train_dataset = save['train_dataset']
  train_labels = save['train_labels']
  valid_dataset = save['valid_dataset']
  valid_labels = save['valid_labels']
  test_dataset = save['test_dataset']
  test_labels = save['test_labels']
  del save  # hint to help gc free up memory
  print('Training set', train_dataset.shape, train_labels.shape)
  print('Validation set', valid_dataset.shape, valid_labels.shape)
  print('Test set', test_dataset.shape, test_labels.shape)
```

notMNIST字典

```python
cdic = {"a":0, "b":1, "c":2, "d":3, "e":4, "f":5, "g":6, "h":7, "i":8, "j":9, "remain":10}
cdic_r = {}
for key, value in cdic.items():
    cdic_r[value] = key
```

随机拼接生成新的数据集

```python
target_image_height = 56
target_image_width = 112
max_time_step = 3

def pad_time_step(data, remain):
    for i in range(len(data)):
        data[i] = np.concatenate((data[i],np.ones((max_time_step-len(data[i])),dtype=np.int32)*remain),axis=0)
    return np.array(data)

#数据生成
def image_generate(src_dataset, src_labels, num):
    new_dataset = np.zeros([num, target_image_height, target_image_width],dtype=np.float32)
    new_labels = []
    new_labels_num = np.zeros([num],dtype=np.int32)
    for i in range(num):
        if random.choice([0,1]) == 0: #三个28*28拼接
            index = np.random.randint(0,src_dataset.shape[0], 3) #随机选出3个图片
            images = src_dataset[index]
            label = src_labels[index]
            newimages = [None]*len(images)
            for j in range(len(images)):
                newimages[j] = misc.imresize(images[j],(target_image_height, target_image_width//3))
            concatimage = np.concatenate(newimages,axis=1)
        else:  #两个28*28拼接
            index = np.random.randint(0,src_dataset.shape[0], 2) #随机选出2个图片
            images = src_dataset[index]
            label = src_labels[index]
            newimages = [None]*len(images)
            for j in range(len(images)):
                newimages[j] = misc.imresize(images[j],(target_image_height, target_image_width//2))
            concatimage = np.concatenate(newimages,axis=1)
        new_dataset[i,:,:] = np.concatenate((concatimage,np.zeros((target_image_height,
                                                                   target_image_width-concatimage.shape[1]),dtype=np.int32)),axis=1)
        new_labels.append(label)
        new_labels_num[i] = len(label)
    return new_dataset, pad_time_step(new_labels,cdic["remain"]), new_labels_num  #输入，输出序列，输出序列元素长度
```

```python
train_dataset1, train_labels1, train_labels_num = image_generate(train_dataset, train_labels, 5)
for i in range(5):
    plt.imshow(train_dataset1[i], plt.cm.gray)
    plt.show()
    print(train_labels1[i])
```
